### Gráficos semelhantes das funções de busca binária e linear:

```python
def log2(x):
    return np.log2(x)

def f(x):
    return x

x = np.arange(1, 32, 0.1)
```

### Função de Log na base 2
##### Gráfico semelhante a eficiência da busca binária:

```python
plt.plot(x, log2(x), color='red')
plt.title('f(x) = Log_2x')
plt.grid(True)
plt.show()
```

##### Gráfico semelhante a eficiência da busca linear:

```python
plt.plot(x, f(x))
plt.title('f(x) = x')
plt.grid(True)
plt.show()
```
